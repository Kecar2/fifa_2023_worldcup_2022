# 4. Mejor PLantilla de Cada País

```python
df_best_country_team = df.copy()
df_best_country_team = df_best_country_team.groupby(['Nationality', 'Best Position']).head(2)
```

```python
df_best_country_team[df_best_country_team['Nationality']=='Argentina'].sort_values(by=['Best Position', 'Overall', 'Potential'], ascending=False)
```

```python
def mejor_plantilla(nacionalidad):
    df_best_country_team = df.copy()
    df_best_country_team = df_best_country_team.groupby(['Nationality', 'Best Position']).head(2)
    df_best_country_team = df_best_country_team[df_best_country_team['Nationality']==nacionalidad].sort_values(by=['Best Position', 'Overall', 'Potential'], ascending=False)
    return df_best_country_team
```

```python
mejor_plantilla('Spain')
```

```python
promedio_puntuacion = [mejor_plantilla(team)['Overall'].mean() for team in teams_worldcup_22]
promedio_puntuacion
```

```python
df_promedio_puntuacion = pd.DataFrame({'Teams': teams_worldcup_22, 'Media_Puntuacion': promedio_puntuacion})
```

# 5. Mejor 11 de Cada Equipo

```python
dict_formation = {
    '4-3-3': ['GK', 'RB', 'CB', 'CB', 'LB', 'CDM', 'CM', 'CAM', 'RW', 'ST', 'LW'],
    '4-4-2': ['GK', 'RB', 'CB', 'CB', 'LB', 'RM', 'CM', 'CM', 'LM', 'ST', 'ST'],
    '4-2-3-1': ['GK', 'RB', 'CB', 'CB', 'LB', 'CDM', 'CDM', 'CAM', 'CAM', 'ST'],
    '3-4-3': ['GK', 'CB', 'CB', 'CB', 'RM', 'CM', 'CM', 'LM', 'RW', 'ST', 'LW'],
    '4-1-2-1-2': ['GK', 'RB', 'CB', 'CB', 'LB', 'CDM', 'RM', 'CAM', 'LM', 'ST', 'ST'],
    '4-2-4': ['GK', 'RB', 'CB', 'CB', 'LB', 'CM', 'CM', 'RW', 'LW', 'ST', 'ST'],
    '5-2-2-1': ['GK', 'RWB', 'CB', 'CB', 'CB', 'LWB', 'CM', 'CM', 'RW', 'ST', 'LW']
}
```

```python
def best_lineup(nationality, lineup):
    lineup_count = [lineup.count(i) for i in lineup]
    
    df_lineup = pd.DataFrame({'position': lineup, 'count': lineup_count})
    positions_non_repeated = df_lineup[df_lineup['count'] <= 1]['position'].values
    positions_repeated = df_lineup[df_lineup['count'] > 1]['position'].values
    
    df_squad = mejor_plantilla(nationality)
    
    df_lineup = pd.concat([
        df_squad[df_squad['Best Position'].isin(positions_non_repeated)].drop_duplicates('Best Position', keep='first'),
        df_squad[df_squad['Best Position'].isin(positions_repeated)]])
    
    return df_lineup[['Known As', 'Overall', 'Club Name', 'Best Position']]
```

```python
best_lineup('Brazil', dict_formation['4-3-3'])
```

```python
for index, row in df_promedio_puntuacion[:9].iterrows():
    max_average = None
    for key, values in dict_formation.items():
        average = best_lineup(row['Teams'], values)['Overall'].mean()
        if max_average is None or average>max_average:
            max_average = average
            formation = key
    print(row['Teams'], formation, max_average)
```
